```python
import os
import json
import pandas as pd

folder_path = "/content/sample_data/crisisfacts_001"
data_list = []
for filename in os.listdir(folder_path):
    if filename.endswith(".json"):
        file_path = os.path.join(folder_path, filename)
        with open(file_path,'r') as json_file:
            json_objects_string = json_file.read()
            json_objects = json_objects_string.strip().split('}{')
            # Add back the missing braces to each object
            for i in range(len(json_objects)):
                if i == 0:
                    json_objects[i] += '}'
                elif i == len(json_objects) - 1:
                    json_objects[i] = '{' + json_objects[i]
                else:
                    json_objects[i] = '{' + json_objects[i] + '}'

            for json_object in json_objects:
                json_data = json.loads(json_object)
                data_list.append(json_data)

news_dataframe = pd.json_normalize(data_list)
print(data_list)
print(news_dataframe)
```

```python
def get_question_context(question_embeddings):
    cosine_values = []
    for text in embedding_dict:
        value = np.dot(question_embeddings, np.array(embedding_dict[text]))
        cosine_values.append({"text": text, "value": value})

    sorted_cosine_values = sorted(cosine_values, key=lambda x: x['value'], reverse=True)

    # Get context string
    context = ""
    counter = 0
    for val in sorted_cosine_values:
        context += val['text'] + "\n"
        counter += 1
        if counter == 15:
            break
    return context    

openai.api_key = "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"
def get_answers(question, context):
    model_engine = "text-davinci-002"
    prompt_template = "Question: {question}\nContext: {context}\nAnswer:"

    prompt = prompt_template.format(question=question, context=context)
    response = openai.Completion.create(
                engine=model_engine,
          prompt=prompt, max_tokens=1024, n=1,
            stop=None,
            temperature=0.5,
        )
    answer = response.choices[0].text.strip()
    return answer

# for query in test_queries:
#     question = query['question']
# question = queries_crisisfact001_dataframe.iloc[12]['query']
# question_embeddings = sentence_model.encode([question])[0]

# context = get_question_context(question_embeddings)
# answer = get_answers(question, context)
# print("Question: ", question)
# print("Answer: ", answer)

```
